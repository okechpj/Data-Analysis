# Axes labels, Legend and Grid

```python
days =  [1,2,3,4,5,6,7]
max_t = [50,53,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label="max temp")
plt.plot(days, min_t, label="min temp")
plt.plot(days, avg_t, label="avg temp")

plt.legend(loc="lower left", shadow=True)

plt.grid()
```

# Bar chart

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
```

```python
plt.ylabel("Revenue(Billions)")
plt.title("US tech stockst")

plt.bar(company, revenue, label="Revenue")

plt.legend()
```

Display more than one data items on a single bar

```python
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40, 2,34,12]
```

```python
plt.ylabel("Revenue(Billions)")
plt.title("US tech stockst")

plt.bar(company, revenue, label="Revenue")
plt.bar(company, profit, label="Profit")

plt.legend()
```

Dislay bars side by side

```python
#import numpy
import numpy as np
```

```python
#Data
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40, 2,34,12]
```

```python
# Include axis labels
plt.ylabel("Revenue(Billions)")
plt.title("US tech stockst")

#set the positions of the bars
bar_width = 0.35
index = np.arange(len(company))

#Plot the graph
plt.bar(index, revenue, width=bar_width, label="Revenue")
plt.bar(index + bar_width, profit, width=bar_width, label="Profit")

#include a legend
plt.legend()

plt.savefig('bar_chart.png', pad_inches=3)
```

# Horizontal bar graphs

```python
#Data
company = ['GOOGLE', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40, 2,34,12]
```

```python
# Include axis labels
plt.ylabel("Revenue(Billions)")
plt.title("US tech stockst")

#set the positions of the bars
bar_width = 0.35
index = np.arange(len(company))

#Plot the graph
plt.barh(index, revenue, height=bar_width, label="Revenue")
plt.barh(index + bar_width, profit, height=bar_width, label="Profit")

#include a legend
plt.legend()
```
